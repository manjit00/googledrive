### Matrix multiplication

```python
tensor_A = torch.tensor([[1,2],
                       [3,4],
                       [5,6]])
tensor_B = torch.tensor([[7,9],
                         [10,11],
                         [12, 13]])
torch.mm(tensor_A.T,tensor_B)

tensor_A.shape, tensor_A.T.shape, tensor_B.shape
## Matrix rule
## the inner dimensions should match
## the resulting matrix will have shape of outer matrixes dimensions
## 3x2, 2x3 = 3x3
## 4x2, 2x5 = 4x5
```

## Reproducibility - taking random out of random.

#### random seed, can be used to lessen the randomness

```python
random_A = torch.rand(3,4)
random_B = torch.rand(3,4)
print(random_A)
print(random_B)
print(random_A == random_B)
```

```python
torch.manual_seed(42)
random_C = torch.rand(3,4)
torch.manual_seed(42)
random_D = torch.rand(3,4)

print (random_C)
print (random_D)
print (random_C == random_D)
```
